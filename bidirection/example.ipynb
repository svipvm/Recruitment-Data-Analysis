```python
import networkx as nx
from networkx.algorithms import bipartite
import numpy as np
import matplotlib.pyplot as plt
```

```python
job_ids = ['job#{}'.format(i) for i in range(4)]
hunter_ids = ['hunter#{}'.format(i) for i in range(5)]
job_ids = list(reversed(job_ids))
hunter_ids = list(reversed(hunter_ids))

edges = [(np.random.choice(job_ids), np.random.choice(hunter_ids)) for _ in range(8)]

binet = nx.Graph()
binet.add_nodes_from(job_ids, bipartite=0)
binet.add_nodes_from(hunter_ids, bipartite=1)
binet.add_edges_from(edges)
# nx.is_connected(binet)

pos = {}
# Update position for node from each group
pos.update((node, (1, index)) for index, node in enumerate(job_ids))
pos.update((node, (2, index)) for index, node in enumerate(hunter_ids))

# nx.draw(binet, pos=pos, with_labels=True)
nx.draw_networkx(binet, pos, with_labels=True)
plt.show()
```

```python
job_ids = ['job#{}_{}'.format(i, j) for i in range(3) for j in range(np.random.randint(1, 3))]
hunter_ids = ['hunter#{}'.format(i) for i in range(5)]
edges = [(np.random.choice(job_ids), np.random.choice(hunter_ids), {'weight': np.random.randint(30, 60)}) for _ in range(8)]
job_ids = list(reversed(job_ids))
hunter_ids = list(reversed(hunter_ids))
# print(edges)

binet = nx.Graph()
binet.add_nodes_from(job_ids, bipartite=0)
binet.add_nodes_from(hunter_ids, bipartite=1)
binet.add_edges_from(edges)
weights = nx.get_edge_attributes(binet, "weight")
# print(weights)
# nx.is_connected(binet)

pos = {}
# Update position for node from each group
pos.update((node, (1, index)) for index, node in enumerate(job_ids))
pos.update((node, (2, index)) for index, node in enumerate(hunter_ids))

nx.draw_networkx(binet, pos, with_labels=True)
nx.draw_networkx_edge_labels(binet, pos, edge_labels=weights)
# nx.draw(binet, pos=pos, with_labels=True)
plt.show()
```
