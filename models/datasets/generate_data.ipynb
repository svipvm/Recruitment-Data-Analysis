```python
import os, random
```

```python
DATASETS_ROOT_DIRS = [dir_ for dir_ in os.listdir('./') if os.path.isdir(dir_)]
DATASETS_ROOT_DIRS
```

```python
def check_text(text_a, text_b):
    if len(text_a) < 4 or len(text_b) < 4: return False
    
    return True
```

```python
head_row = '\t'.join(['text_a', 'text_b', 'label\n'])
data_lines = {'train': [head_row], 'dev': [head_row]}
contain_map = {'entailment': "0\n", 'neutral': "1\n", 'contradiction': "2\n"}

for root_dir in DATASETS_ROOT_DIRS:
    files = os.listdir(root_dir)
    assert any(['train.txt' in file or 'dev.txt' in file for file in files])
    if root_dir in ['CINLID', 'ChineseTextualInference', 'OCNLI', 'CSNLI', 'CMNLI']:
        for data_type in ['train', 'dev']:
            with open(f'{root_dir}/{data_type}.txt', 'r', encoding='UTF-8') as f:
                all_lines = f.readlines()
                for line in all_lines:
                    line_ = line.replace('\n', '').split('\t')
                    if len(line_) != 3: 
                        print(root_dir, line)
                        continue
                    if not check_text(line_[0], line_[1]): continue
                    line_[2] = contain_map[line_[2]]
                    data_lines[data_type].append('\t'.join(line_))
    elif root_dir in ['AFQMC', 'LCQMC', 'OPPO-xiaobu', 'PAWS-X']:
        for data_type in ['train', 'dev']:
            with open(f'{root_dir}/{data_type}.txt', 'r', encoding='UTF-8') as f:
                all_lines = f.readlines()
                for line in all_lines:
                    line_ = line.replace('\n', '').split('\t')
                    if len(line_) != 3: 
                        print(root_dir, line)
                        continue
                    if not check_text(line_[0], line_[1]): continue
                    line_[2] = "0\n" if int(line_[2]) == 1 else '2\n'
                    data_lines[data_type].append('\t'.join(line_))
    elif root_dir in ['Chinese-STS-B']:
        for data_type in ['train', 'dev']:
            with open(f'{root_dir}/{data_type}.txt', 'r', encoding='UTF-8') as f:
                all_lines = f.readlines()
                for line in all_lines:
                    line_ = line.replace('\n', '').split('\t')
                    if len(line_) != 3: 
                        print(root_dir, line)
                        continue
                    if not check_text(line_[0], line_[1]): continue
                    line_[2] = str(2 - int(line_[2]) // 2) + '\n'
                    data_lines[data_type].append('\t'.join(line_))
    elif root_dir in ['PKU-Paraphrase-Bank']:
        for data_type in ['train', 'dev']:
            with open(f'{root_dir}/{data_type}.txt', 'r', encoding='UTF-8') as f:
                all_lines = f.readlines()
                all_lines = all_lines[::20]
                for line in all_lines:
                    line_ = line.replace('\n', '').split('\t')
                    if len(line_) != 2: 
                        print(root_dir, line)
                        continue
                    if not check_text(line_[0], line_[1]): continue
                    line_.append('0\n')
                    data_lines[data_type].append('\t'.join(line_))
```

```python
print('train:', len(data_lines['train']), 'dev:', len(data_lines['dev']))

with open('train.tsv', 'w', encoding='UTF-8') as f:
    f.writelines(data_lines['train'])

with open('dev.tsv', 'w', encoding='UTF-8') as f:
    f.writelines(data_lines['dev'])
```
